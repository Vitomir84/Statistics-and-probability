### Derivative Rules
We have seen how to calculate the derivative of a couple of functions. But how
should we proceed when, for instance, the function we want to differentiate is
made of multiple functions, multiplied or interleaved?
#### MULTIPLICATION BY CONSTANT
Constants can be extracted from derivatives:
$$
\frac{dαf(x)}
{dx} = α \frac{df(x)}{dx}
$$


#### FUNCTION ADDITION: THE SUM RULE
The derivative of the sum of functions is equivalent to the sum of the derivatives
of these functions:
$$
\frac{df(x)+g(x)}{dx} = \frac{df(x)}{dx} + \frac{dg(x)}{dx}
$$
Using the sigma notation we can write:
For instance,
$$
\frac{d\sum{f(x)}}{dx} = \sum{\frac{dx}{dx}}
$$

#### FUNCTION MULTIPLICATION: THE PRODUCT RULE
When the function you want to differentiate is made of the multiplication of two
functions, we can use the product rule. Let’s say that we have two functions, f(x)
and g(x). With Leibniz notation, we write their derivatives as $\frac{df(x)}{dx}$ and $\frac{dg(x)}{dx}$
respectively. Remember that $\frac{df(x)}{dx}$ just refers to the derivative of f(x) in a
function of x. 
The function we want to differentiate is made of the multiplication of and
thus, we can write its derivative: 
$$
\frac{df(x)(g(x)}{dx} = \frac{df(x)}{dx} * g(x) + f(x)*\frac{dg(x)}{dx}
$$

#### FUNCTION COMPOSITION: THE CHAIN RULE
Composite function refers to the composition of two functions: one of them that
takes the result of the other one as input. The concepts are similar in mathematics
and in computer science. For instance, let’s consider the following functions:
$$
h(x) = f (g(x)) = f (2x) = (2x)^2
$$

```python
def f(x):
    print(f"g(x): input: {x}, output: {x} ** 2 = {x ** 2}")
    return x ** 2

def g(x):
    print(f"f(x): input: {x}, output: 2 * {x} = {2 * x}")
    return 2 * x
```

```python
def h(x):
    return f(g(x))
```

### Izvod cost funkcije

$$
J(θ_1) = \frac{1}{2m}\sum_{m}^{i} (θ_1 x^i) - y^i)^2
$$

Let’s differentiate this cost function $J(θ_1)$. It can be written $\frac{d J(θ_1)}{dx}$, meaning that the
differentiation is done with respect to $θ_1$.

$$
\frac{d J(θ_1)}{d θ_1 } = \frac{d}{d θ_1 } (\frac{1}{2m}\sum_{m}^{i}(θ_1x^i - y^i)^2)
$$


We will use two rules from rules above. We can get out from the
derivative because of the multiplication by constant rule. The sum can also be
extracted from the derivative using the sum rule.

$$
\frac{d J(θ_1)}{d θ_1 } =  \frac{1}{2m}\sum_{m}^{i} \frac{d}{d θ_1 }(θ_1x^i - y^i)^2
$$

To differentiate this function, we need to separate it as the combination of two
functions. So, let’s create the following function: 

$$
u(θ_1 ,i) = θ_1x^{i} −y^{i}
$$

We can replace this in our cost function:
Using the chain rule, we can find that the derivative of the cost function is the
following:

$$
J(θ_1) =  \frac{1}{2m}\sum_{m}^{i}(u(θ_1 ,i))^2
$$


Let’s calculate derivatives of and separately:

$$
\frac{d J(θ_1)}{d u } = \frac {d \frac{1}{2m}\sum_{m}^{i}(u(θ_1 ,i))^2}{du}
$$


Note that this derivative is calculated with respect to the function u. Then we
have:

$$
\frac{du}{d θ_1 } = \frac {d (θ_1x^i - y^i)}{d θ_1}
$$

Now that we have calculated the derivative of both functions, we can use the
chain rule:

$$
\frac{dJ}{d θ_1 } = \frac{1}{2m} \sum_{m}^{i}2u(θ_1 * x^i)
$$

And replacing back into the equation, we have:

$$
\frac{dJ}{d θ_1 } = \frac{1}{2m} \sum_{m}^{i}2(θ_1*x^i - y^i)*x^i
$$


The constant goes out from the sum and we have our derivative:

$$
\frac{dJ}{d θ_1 } = \frac{1}{m} \sum_{m}^{i}(θ_1*x^i - y^i)*x^i
$$

We have calculated the derivative of the MSE cost function. This function $\frac{dJ}{d θ_1 }$ is
helpful because it takes a value of the parameter $θ_1$ we want to optimize and
returns the slope of the tangent of the cost function for this value of $θ_1$. The slope
tells us the direction to take to minimize the cost.


```python
def calculate_error_j_(theta1, x_i, y_i):
    return ((theta1 * x_i) - y_i) * x_i
```
