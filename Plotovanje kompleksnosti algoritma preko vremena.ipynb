# Plotovanje vremena izvr≈°enja komande u zavisnosti od kompleksnosti algoritma

```python
import time
import math
from matplotlib import pylab as plt
import seaborn as sns
```

```python
linearna_kompleksnost=[]
eksponencijalna_kompleksnost=[]
logaritamska_kompleksnost=[]
```

```python
for n in range(1,10000):
    start=time.time()
    zbir=n+n
    end=time.time()
    linearna_kompleksnost.append(end-start)
for n in range(1,10000):
    start=time.time()
    stepen=n**n
    end=time.time()
    eksponencijalna_kompleksnost.append(end-start)
for n in range(1,10000):
    start=time.time()
    logaritam=math.log2(n)
    end=time.time()
    logaritamska_kompleksnost.append(end-start)
```

```python
fig, ax = plt.subplots(1, figsize=(20, 10))
ax.plot(linearna_kompleksnost, label='Vreme potrebno da se izraƒçuna n+n O(n)', color='red')
ax.plot(eksponencijalna_kompleksnost, label='Vreme potrebno da se izraƒçuna n2 (On2)', color='blue')
ax.plot(logaritamska_kompleksnost, label='Vreme potrebno da se izraƒçuna log(n) (Olog(n))', color='green')
plt.xlabel('n')
plt.ylabel('vreme')
ax.legend()
plt.show()
```
